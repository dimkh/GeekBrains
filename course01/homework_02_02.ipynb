```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-5, 5, 21)
y = 3*x + 1
y2 = (-1/3)*x + 1
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title('Выровнен масштаб осей x и y')
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.gca().set_aspect('equal', adjustable='box')
```
