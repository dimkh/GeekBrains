```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

x = np.linspace(-2, 5, 201)
plt.plot(x, 1 - (( 1 - np.exp(x) )/x))
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,20) 
plt.grid(True)
plt.show()

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)

x1, y1 = fsolve(equations, (-2, 0))
x2, y2 = fsolve(equations, (2, 3))
x3, y3 = fsolve(equations, (3.5, 4.5))
print (x1, y1)
print (x2, y2)
print (x3, y3)
```
