```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = []
y1 = []
y2 = []
a = 1
b = 2

a = a**2
b = b**2
step = 0.05

#Рассчитываем левую часть графика
t = -a-5
while t <= -a:
    x_ = t
    x.append(x_)
    y_ = math.sqrt( ((x_**2 / a) - 1)*b)
    y1.append(y_)
    y2.append(-y_)
    t += step
#Рассчитываем правую часть графика
t = a
while t < 5+a:
    x_ = t
    x.append(x_)
    y_ = math.sqrt( ((x_**2 / a) - 1)*b)
    y1.append(y_)
    y2.append(-y_)
    t += step

plt.plot(x, y1, color = 'blue')
plt.plot(x, y2, color = 'blue')
plt.plot([-a-5,5+a], [0,0], color = 'black')
plt.plot([0,0], [y2[0],y1[(len(y1)-1)]], color = 'black')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title('График гиперболы')
```
