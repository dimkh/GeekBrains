```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-2*np.pi, 3*np.pi, 201)
k1 = 0.5
k2 = 1
k3 = 2
a1 = 1
a2 = 20
a3 = 45
b1 = 0
b2 = 1
b3 = -1
plt.plot(x, k1*np.cos(x - a1) + b1)
plt.plot(x, k2*np.cos(x - a2) + b2)
plt.plot(x, k3*np.cos(x - a3) + b3)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title('График функций y(x) = k*cos(x-a)+b')
plt.show()
```
