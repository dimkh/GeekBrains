```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.subplot(111, polar=True)        # Полярная система координат
r = 5

fi = np.arange(0, 2*np.pi, 0.01)
rho = r*fi/fi
plt.plot(fi, rho, lw=2)
plt.title(f'График окружности радиуса {r} в полярных координатах')
plt.show()
```
