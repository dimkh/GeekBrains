```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Центр окружности (A,B), радиус R
A = 3
B = 2
R = 3
step = 0.05

x = []
y = []
t = 0
# Вычисляем точку окружности для угла между 0 и 2пи
while t < 2 * np.pi:
    x.append(R * np.cos(t) + A)
    y.append(R * np.sin(t) + B)
    t += step

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title(f'Окружность радиусом {R} с центром в ({A},{B})')
plt.xlim(A-R-0.5, A+R+0.5)
plt.ylim(B-R-0.5, B+R+0.5)
plt.gca().set_aspect('equal', adjustable='box')
```
