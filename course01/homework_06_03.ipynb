```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
B = [[1, 2, 3, 12], [4, 5, 6, 2], [7, 8, 9, 1]]
B1 = [[1, 2, 3, 6], [4, 5, 6, 24], [7, 8, 9, 42]]

r_A = np.linalg.matrix_rank(A, 0.0001)
r_B = np.linalg.matrix_rank(B, 0.0001)
r_B1 = np.linalg.matrix_rank(B1, 0.0001)

print(f'Ранг матрицы A равен: {r_A}')
print(f'Ранг матрицы B равен: {r_B}')
print()
print('Ранг матрицы A не совпадает с рангом расширенной матрицы B (он меньше),')
print('поэтому данная система уравнений не имеет ни одного решения.')
print()
print('Изменим правую часть - сделаем её линейно-зависимой первому столбцу:')
print(B1[0])
print(B1[1])
print(B1[2])
print()
print(f'Ранг матрицы A равен: {r_A}')
print(f'Ранг матрицы B1 равен: {r_B1}')
print()
print('Ранг матрицы A (исходной) теперь совпадает с рангом матрицы B1 (расширенной),')
print('но ранг меньше количества неизвестных. Данная система имеет бесчисленное множество решений.')
print()
print('Чтобы система имела одно решение, необходимо, чтобы ранги исходной и расширенной матриц')
print('были равны и равнялись количеству неизвестных. Но по условию задачи мы меняем только правую часть -')
print('расширенную матрицу.')

```
