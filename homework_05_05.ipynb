```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math

n = 100
r = 0.7
x = np.random.rand(n)
y = r*x + (1 - r)*np.random.rand(n)
plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

a = (np.sum(x)*np.sum(y) - n*np.sum(x*y))/(np.sum(x)*np.sum(x) - n*np.sum(x*x))
b = (np.sum(y) - a*np.sum(x))/n
avg_x = np.mean(x)
avg_y = np.mean(y)
c = np.sum((x-avg_x)*(y-avg_y)) / math.sqrt(sum((x-avg_x)**2)*sum((y-avg_y)**2))

A = np.vstack([x, np.ones(len(x))]).T
a1, b1 = np.linalg.lstsq(A, y)[0]
c1 = np.corrcoef(x, y) 
print(f'Расчёт коэффициентов уравнения прямой вручную:        a = {a}, b = {b}')
print(f'Расчёт коэффициентов уравнения функцией linalg.lstsq: a = {a1}, b = {b1}')
print()
print(f'Коэффициент корреляции (ручной расчёт) c = {c}')
print(f'Коэффициент корреляции (функция corrcoef) c = {c1}')
plt.plot([0, 1], [b, a + b])
plt.show()
```
