```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X2 = np.arange(-2,2,0.1)
Y2 = np.arange(-2,2,0.1)
X2,Y2 = np.meshgrid(X2,Y2)

def f2(x,y):
    return ((x**2 / 2 - y**2 / 3))

ax.plot_surface(X2,Y2,f2(X2,Y2))
plt.title('Гиперболический параболоид')
plt.show()
```
