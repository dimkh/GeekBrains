```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Формируем 10 массивов по 10 случайных чисел в диапазоне от 0 до 9 включительно
arr = []
sum_elem = []
for i in range(0,10):
    arr.append(np.random.randint(0,9,10))
    sum_elem.append(sum(arr[i]))            # Сразу же считаем сумму элементов каждого массива

num_bins = 5            # Разделим всю выборку на 5 диапазонов
all_range = (0,100)     # Диапазон сумм - от 0 до 100
plt.hist(sum_elem, num_bins, all_range, color = 'darkblue', histtype = 'bar', rwidth = 0.95)
plt.xlabel('Диапазон суммы')
plt.ylabel('Количество')
plt.title('Распределение средней суммы')
plt.show()
```
