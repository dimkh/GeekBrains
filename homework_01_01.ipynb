```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-5, 5, 401)
k = 2
plt.plot(x, np.cos(x))
plt.plot(x, np.cos(x*k))
plt.title('График функций y(x) = cos(k*x)', fontsize=20, fontname='Times New Roman')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.grid(True)
plt.legend(['k=1', 'k=2'])
plt.show()
```
