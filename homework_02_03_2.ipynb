```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Центр эллипса (A,B)
A = 2
B = 1
# Полуоси эллипса
axa = 3
axb = 2
step = 0.05

x = []
y = []
t = 0
# Вычисляем точку эллипса для угла между 0 и 2пи
while t < 2 * np.pi:
    x.append(axa * np.cos(t) + A)
    y.append(axb * np.sin(t) + B)
    t += step

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title(f'Эллипс с центром в ({A},{B}) и полуосями {axa}, {axb}')
plt.xlim(A-axa-0.5, A+axa+0.5)
plt.ylim(B-axb-0.5, B+axb+0.5)
plt.gca().set_aspect('equal', adjustable='box')
```
