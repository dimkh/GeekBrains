```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X1 = np.arange(-2,2,0.1)
Y1 = np.arange(-2,2,0.1)
X1,Y1 = np.meshgrid(X1,Y1)

def f1(x,y):
    return ((x**2 / 2 + y**2 / 3))

ax.plot_surface(X1,Y1,f1(X1,Y1))
plt.title('Эллиптический параболоид')
plt.show()
```
