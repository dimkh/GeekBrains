```python
# Задание 2

t = np.linspace(0, 10, 51)

f = np.cos(t)

plt.plot(t, f, color = 'green')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

```python
# Задание 3

x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)
ax1.set_title('График $y_1$')
ax2.set_title('График $y_2$')
ax3.set_title('График $y_3$')
ax4.set_title('График $y_4$')
ax1.set_xlim([-5, 5])
fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
plt.show()
```

```python
# Задание 4

import pandas as pd

# По условиям задания нужно подключить стиль 'fivethirtyeight': подключил и закомментировал.
# Так как использую тёмную тему, то этот стиль не очень подходит - почти не видно осей, значений и т.п.
# Оставляю тёмный стиль из заданий выше.
# plt.style.use('fivethirtyeight')
cards = pd.read_csv('creditcard.csv')
```

```python
class_lst = cards['Class'].value_counts()
print(f'Количество обычных транзакций {class_lst[0]}, мошеннических - {class_lst[1]}')

class_lst.plot(kind = 'bar')
plt.title('Число обычных и мошеннических транзакций')
# plt.savefig('picture_1', dpi = 300)
plt.show()
# И второго столбика не видно, т.к. он почти в 600 раз меньше размером.
```

```python
class_lst.plot(kind = 'bar', logy = True)
plt.title('Число обычных и мошеннических транзакций')
# plt.savefig('picture_2', dpi = 300)
plt.show()
# На логарифмическом масштабе видны оба столбика.
```

```python
transact_norm = cards.loc[cards['Class'] == 0, ['V1']]
transact_fraud = cards.loc[cards['Class'] == 1, ['V1']]

# По условиям задачи гистограмма для обычных транзакций должна быть серого цвета
# Синий на тёмном фоне выглядит более предпочтительным

plt.hist(transact_norm['V1'], bins = 20, density = True, alpha = 0.5, label = 'Normal transactions', color = 'blue')
plt.hist(transact_fraud['V1'], bins = 20, density = True, alpha = 0.5, label = 'Fraud transactions', color = 'red')
plt.title('Распределение параметра V1')
plt.legend()
# plt.savefig('picture_3', dpi = 300)
plt.show()
```
