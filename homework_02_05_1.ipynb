```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
#print(X)
X, Y = np.meshgrid(X, Y)
# print(Y)
# print(X[1,1])
Z = X + 2*Y
Z2 = X + 2*Y + 30
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```
