```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.subplot(111, polar=True)        # Полярная система координат
rho1 = 30
fi1 = 0     # задано в градусах
rho2 = 10
fi2 = 90    # задано в градусах

# Переводим в радианы
fi1 = fi1 * np.pi / 180
fi2 = fi2 * np.pi / 180

fi = np.linspace(fi1, fi2, num=100, endpoint=True)
rho = np.linspace(rho1, rho2, num=100, endpoint=True)

plt.plot(fi, rho, lw=2)
plt.title(f'График отрезка ({rho1},0) (0,{rho2}) в полярных координатах')
plt.show()
```
