```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

print('Две плоскости пересекаются по прямой.')
print('Система заданных линейных уравнений её определяет.')
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.1)
Y = np.arange(-5,5,0.1)
X, Y = np.meshgrid(X, Y)
Z1 = X + 2*Y - 1
Z2 = -4*X + 5/2*Y + 6
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
show()

print('')
print('Из первого уравнения выразим z и подставим во второе уравнение.')
print('z = x + 2*y - 1')
print('8*x - 5*y + 2*x + 4*y - 2 - 12 = 0')
print('10*x - y - 14 = 0')
print('Полученное уравнение: y = 10*x - 14')
print('')
print('Найдём минимальное расстояние до нулевого вектора (графически):')
print('')

def Q(x, y):
    return (x**2 + y**2)
    
x = np.linspace(1.3, 1.5, 201)
plt.plot(x, Q(x, 10 * x - 14))
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

print('')
print('Видно, что x приблизительно равен 1.382')
print('')
print('Вычислим остальные значения:')
print('y = 10*1.382 - 14 = -0.18')
print('z = 1.382 - 2*0.18 - 1 = 0.022')
print('')
print('Итак, нашли, что x = 1.382, y = -0.18, z = 0.022')
print('')
print('Расчёт средствами Python:')

A = [[1, 2, -1], [8, -5, 2]]
B = [1, 12]
print(np.linalg.lstsq(A, B))
print('')
print('Получили практически те же самые значения.')
```
